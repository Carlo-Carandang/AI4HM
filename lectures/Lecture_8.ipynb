## $k$-Means

* $$\text{Minimize } \sum_{\ell=1}^k W(C_k)$$
* $$\text{subject to } \bigcup_{\ell=1}^k C_k = [n] \text{ and } C_\ell \cap C_{\ell'} = \emptyset;$$
* $$\text{for } W(C) = \frac{1}{|C|} \sum_{i,j\in C} \|x_i-x_j\|^2.$$

```python
# You try it
df = pd.read_csv('/home/j/Project/Machine-Learning/ML4HM/ISL/Fig_2_9_data.csv', index_col=0)
X = array(df.filter(like='X'))
labels = cluster.KMeans(n_clusters=2).fit_predict(X)
```
