First, an example of k-NN in sklearn

```python
import sklearn.datasets, sklearn.neighbors
```

```python
iris = sklearn.datasets.load_iris()
print iris.DESCR
```

```python
X = iris.data[:, :2]  # we only take the first two features. We could
                      # avoid this ugly slicing by using a two-dim dataset
y = iris.target
```

```python
n_neighbors=15
weights='uniform'
clf = sklearn.neighbors.KNeighborsClassifier(n_neighbors, weights)
clf.fit(X, y)
```

```python
# explore the decision boundary. For that, we will predict each
# point in the mesh [x_min, m_max]x[y_min, y_max].
h = .02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Create color maps
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
# Put the result into a color plot
figure(figsize=(8,5))

Z = Z.reshape(xx.shape)
pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
title("3-Class classification (k = %i, weights = '%s')"
             % (n_neighbors, weights))

axis('tight')
pass
```

## Now to reproduce figure 2.15

```python
import pandas as pd
df = pd.read_csv('/home/j/Project/Machine-Learning/ML4HM/ISL/Fig_2_9_data.csv', index_col=0)
```

```python
# fix the color names

df['color'] = df['Y'].map({'red': 'orange', 'blue': 'purple'})
```

```python
# make Figure 2.9

figure(figsize=(8,8))
for g, dfg in df.groupby('color'):
    plot(dfg['X.1'], dfg['X.2'], 'o', mec=g, color='none', mew=3, ms=8)
```

```python
n_neighbors=10
weights='uniform'
X = array(df.filter(like='X'))
y = array(df.color.map({'orange':0.0, 'purple':1.0}))
clf = sklearn.neighbors.KNeighborsClassifier(n_neighbors, weights)
clf.fit(X, y)
```

```python
# explore the decision boundary. For that, we will predict each
# point in the mesh [x_min, m_max]x[y_min, y_max].
h = .02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
figure(figsize=(8,8))

# show k-NN decision boundary
Z = Z.reshape(xx.shape)
contour(xx, yy, Z, levels=[.5], colors=['k'], linewidths=[2])

# also the training points
for g, dfg in df.groupby('color'):
    plot(dfg['X.1'], dfg['X.2'], 'o', mec=g, color='none', mew=3, ms=8, zorder=-10)
    
title("KNN (k = %i, weights = '%s')" % (n_neighbors, weights))

axis('tight')
pass
```

```python
import pymc as mc
```

```python
# also get the bayesian decision boundary, if possible
blue_mean = pd.read_csv('ISL/Fig_2_9_blue_mean.csv', index_col=0)
orange_mean = pd.read_csv('ISL/Fig_2_9_orange_mean.csv', index_col=0)
```

```python
def pr_orange(y):
    like_blue = [mc.normal_like(y, X_j, .5**-2) for j, X_j in blue_mean.iterrows()]
    like_orange = [mc.normal_like(y, X_j, .5**-2) for j, X_j in orange_mean.iterrows()]
    
    return exp(mc.flib.logsum(like_orange) - mc.flib.logsum(like_blue + like_orange))
pr_orange([0,0])
```

```python
h = .1

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = pd.DataFrame([xx.ravel(), yy.ravel()]).apply(pr_orange)
```

```python
# Put the result into a color plot
figure(figsize=(8,8))

# show k-NN decision boundary
Z = Z.reshape(xx.shape)
contour(xx, yy, Z, levels=[.5], colors='purple', linewidths=[2], linestyles='dashed')

# also the training points
for g, dfg in df.groupby('color'):
    plot(dfg['X.1'], dfg['X.2'], 'o', mec=g, color='none', mew=3, ms=8, zorder=-10)
    
title("KNN (k = %i, weights = '%s')" % (n_neighbors, weights))

axis('tight')
pass
```
